**finding token lengths of test file**

```python

import json
from transformers import LlamaTokenizer

# Initialize the LLaMa tokenizer
tokenizer = LlamaTokenizer.from_pretrained('TheBloke/Llama-2-7B-GPTQ')

# Path to the JSONL file
folder_name = '/data/chq'
test_file_name = 'test.jsonl'

# Join the folder and file names to create a full path
jsonl_file_path = os.path.join(folder_name, test_file_name)


# Function to read JSONL file
def read_jsonl(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield json.loads(line)

# Function to calculate tokens and their length
def calculate_token_length(data, tokenizer):
    lengths = []
    for item in data:
        inputs_text = item['inputs']
        tokens = tokenizer.tokenize(inputs_text, is_split_into_words=True)
        num_tokens = len(tokens)
        lengths.append(num_tokens)
        print(f"Inputs: {inputs_text}")
        print(f"Tokens: {tokens}")
        print(f"Number of tokens: {num_tokens}")
        print()
    return lengths

# Read JSONL data
data = list(read_jsonl(jsonl_file_path))

# Calculate token lengths
token_lengths = calculate_token_length(data, tokenizer)

# Path to the output text file
output_file_path = 'token_lengths_words.txt'

# Write the token lengths to the text file in sequential order
with open(output_file_path, 'w') as file:
    for length in token_lengths:
        file.write(f"{length}\n")

print(f"Token lengths have been written to {output_file_path}")
```

# **FINAL TOKENIZER of Source Document of test.jsonl**

```python
import json
from transformers import LlamaTokenizer

# Initialize the LLaMa tokenizer
tokenizer = LlamaTokenizer.from_pretrained('TheBloke/Llama-2-7B-GPTQ')

# Path to the JSONL file
folder_name = '/data/chq'
test_file_name = 'test.jsonl'

# Join the folder and file names to create a full path
jsonl_file_path = os.path.join(folder_name, test_file_name)


# Function to read JSONL file
def read_jsonl(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield json.loads(line)

# Function to calculate tokens and their length
def calculate_token_length(data, tokenizer):
    lengths = []
    max_length=0
    for item in data:
        inputs_text = item['inputs']
        # Split inputs text into words
        words = inputs_text.split()
        # Tokenize each word separately
        tokens = []
        for word in words:
            tokens.extend(tokenizer.tokenize(word))
        num_tokens = len(tokens)
        lengths.append(num_tokens)
        print(f"Inputs: {inputs_text}")
        print(f"Tokens: {tokens}")
        print(f"Number of tokens: {num_tokens}")
        print()
        if num_tokens > max_length:
            max_length = num_tokens
    return lengths, max_length


# Read JSONL data
data = list(read_jsonl(jsonl_file_path))

# Calculate token lengths
token_lengths, max_tok_length = calculate_token_length(data, tokenizer)

# Path to the output text file
output_file_path = 'token_lengths_words.txt'

# Write the token lengths to the text file in sequential order
with open(output_file_path, 'w') as file:
    for length in token_lengths:
        file.write(f"{length}\n")

print(f"Token lengths have been written to {output_file_path}")

print(f"Maximum token length:  {max_tok_length}")
```
